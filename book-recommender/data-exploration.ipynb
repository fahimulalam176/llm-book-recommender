```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd

books = pd.read_csv(path + "/books.csv")

books
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)


plt.xlabel("Column")
plt.ylabel("Missing Values")

plt.show()
```

```python
from datetime import date

# Get the current date
current_date = date.today()

# Access the year attribute to get the current year
current_year = current_date.year
```

```python
import numpy as np

books["missing_Description"] = np.where(books["description"].isna(), 1, 0)  # 1 if true when description is not available, 0 if false.
books["age_of_book"] = current_year - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "average_rating", "ratings_count", "missing_Description", "age_of_book"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 8))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",
                      cbar_kws={"label": "Spearman Correlation"})
plt.show()
```

```python
book_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
].copy()

book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25].copy()
```

```python
book_missing_25_words
```

```python
book_missing_25_words["title and subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)

book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["title", "subtitle"]].astype(str).agg(" ".join, axis=1)
```

```python
book_missing_25_words
```

```python
book_missing_25_words.drop(
    columns=["subtitle", "missing_Description", "age_of_book", "words_in_description"], axis=1
).to_csv("books_cleaned.csv", index=False)
```
